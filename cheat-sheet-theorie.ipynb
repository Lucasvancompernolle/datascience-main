## Formulas for Mean, Variance, and Standard Deviation of a Sample

### Mean (Average) of a Sample
The mean of a sample is the sum of all the values divided by the number of values in the sample.

Formula:
\[
\text{Mean} = \frac{1}{n} \sum_{i=1}^{n} x_i
\]
Where:
- \( n \) is the number of values in the sample.
- \( x_i \) represents each individual value in the sample.

### Variance of a Sample
The variance measures the spread or dispersion of data points in a sample. It's calculated as the average of the squared differences between each data point and the mean.

Formula:
\[
\text{Variance} = \frac{1}{n-1} \sum_{i=1}^{n} (x_i - \text{Mean})^2
\]
Where:
- \( n \) is the number of values in the sample.
- \( x_i \) represents each individual value in the sample.
- \(\text{Mean}\) is the calculated mean of the sample.

### Standard Deviation of a Sample
The standard deviation is the square root of the variance. It provides a measure of how much the data points deviate from the mean.

Formula:
\[
\text{Standard Deviation} = \sqrt{\text{Variance}}
\]
Where:
- \(\text{Variance}\) is the calculated variance of the sample.

# Kansrekening: Basisbegrippen, Standaardnormale Verdeling, en Z-Score

## Basisbegrippen in Kansrekening

1. **Kansruimte (Sample Space)**: De kansruimte, vaak aangeduid als \( S \), is de set van alle mogelijke uitkomsten van een experiment. Bijvoorbeeld, bij het gooien van een munt is de kansruimte \( S = \{Kop, Munt\} \).

2. **Gebeurtenis (Event)**: Een gebeurtenis is een subset van de kansruimte die één of meer mogelijke uitkomsten van het experiment bevat.

3. **Uitkomst (Outcome)**: Een uitkomst is een enkele mogelijke uitkomst van een experiment, en het is een element van de kansruimte.

4. **Waarschijnlijkheidsruimte (Probability Space)**: De waarschijnlijkheidsruimte is een triplet \( (S, F, P) \), waarbij \( S \) de kansruimte is, \( F \) de verzameling van gebeurtenissen is, en \( P \) de kansmaat is.

## Standaardnormale Verdeling

De standaardnormale verdeling, of Z-verdeling, heeft gemiddelde \( \mu = 0 \) en standaardafwijking \( \sigma = 1 \).

**Eigenschappen**:
- Symmetrisch rond \( \mu = 0 \).
- Ongeveer 68% valt binnen 1 \( \sigma \).
- Ongeveer 95% valt binnen 2 \( \sigma \).
- Ongeveer 99.7% valt binnen 3 \( \sigma \).

## Z-Score
 
De Z-score geeft aan hoeveel $ {\sigma} $ een waarde afwijkt van $\mu$ :
$ z = \frac{x - \mu}{\sigma}$

# De centrale limietstelling
De centrale limietstelling (CLT) is een fundamenteel concept in statistiek en waarschijnlijkheidstheorie. Het stelt dat wanneer je een groot genoeg aantal onafhankelijke en identiek verdeelde (i.i.d.) willekeurige variabelen bij elkaar optelt (of hun gemiddelde berekent), de verdeling van de som (of het gemiddelde) asymptotisch normaal wordt, ongeacht de oorspronkelijke verdeling van de variabelen. Met andere woorden, de verdeling van de steekproefgemiddelden nadert naar een normale verdeling, zelfs als de oorspronkelijke verdeling niet normaal is.

# Belang van de Centrale Limietstelling 
De centrale limietstelling is van groot belang omdat het de basis legt voor veel statistische methoden, met name bij steekproefgebaseerde inferentie. Het stelt onderzoekers in staat om de normale verdeling te gebruiken om conclusies te trekken over steekproeven, zelfs als de populatieverdeling niet normaal verdeeld is. Hierdoor kunnen we methoden zoals hypothesetesten en betrouwbaarheidsintervallen gebruiken, die sterk afhankelijk zijn van de eigenschappen van de normale verdeling.



````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Sample data
data = [32.1, 29.9, 31.5, 33.7, 30.4, 31.2, 32.5, 29.8, 30.6, 31.9]

# Confidence level (in decimal)
confidence_level = 0.95

# Calculate the sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = stats.tstd(data)

# Calculate the critical Z-score for the given confidence level
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_critical * (sample_std / np.sqrt(len(data)))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Create a range of x values for plotting
x = np.linspace(sample_mean - 3 * sample_std, sample_mean + 3 * sample_std, 100)

# Calculate the normal distribution values for the x range
y = stats.norm.pdf(x, sample_mean, sample_std)

# Create the plot
plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, y, where=[(x >= confidence_interval[0]) & (x <= confidence_interval[1]) for x in x], color='skyblue', alpha=0.5, label='95% Confidence Interval')
plt.axvline(sample_mean, color='red', linestyle='dashed', label='Sample Mean')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Confidence Interval for Population Mean')
plt.legend()
plt.grid(True)

# Add Z-values at the bottom line of the plot
plt.text(confidence_interval[0], -0.02, f'Z = {z_critical:.2f}', verticalalignment='center', horizontalalignment='center', color='black')
plt.text(confidence_interval[1], -0.02, f'Z = {z_critical:.2f}', verticalalignment='center', horizontalalignment='center', color='black')

# Show the plot
plt.show()
````

# Statistische Toetsen in Python

## Basisbegrippen voor Statistische Toetsen

- **Toets**: Een statistische toets, ook wel hypothesetoets genoemd, is een procedure om te bepalen of er genoeg bewijs is om een bewering over een populatie te aanvaarden of af te wijzen.

- **Nulhypothese (\(H_0\))**: De nulhypothese is de bewering die we testen. Het vertegenwoordigt geen verandering, geen effect, of geen verschil. Bijvoorbeeld: \(H_0: \mu = \mu_0\), waarbij \(\mu\) het populatiegemiddelde is en \(\mu_0\) de waarde in de nulhypothese.

- **Alternatieve Hypothese (\(H_1\) of \(H_a\))**: De alternatieve hypothese is het tegenovergestelde van de nulhypothese. Het vertegenwoordigt de bewering die we proberen te ondersteunen. Bijvoorbeeld: \(H_1: \mu > \mu_0\) (rechtszijdige toets) of \(H_1: \mu \neq \mu_0\) (tweezijdige toets).

- **Steekproefgrootheid**: Een statistische maat die is berekend op basis van de gegevens in de steekproef. Bijvoorbeeld: het steekproefgemiddelde of de steekproefstandaarddeviatie.

- **Significantieniveau (\(\alpha\))**: Het vooraf gekozen kritieke niveau van significantie dat aangeeft hoeveel bewijs er nodig is om de nulhypothese te verwerpen. Typische waarden zijn 0.05 (5%) of 0.01 (1%).

- **Overschrijdingskans (p-waarde)**: De kans om een steekproefgrootheid te verkrijgen die net zo extreem is als (of extremer dan) de waargenomen steekproefgrootheid, gegeven dat de nulhypothese waar is.

- **Aanvaardings- en Kritieke Gebied**: Het aanvaardingsgebied zijn de waarden van de steekproefgrootheid waarin we de nulhypothese accepteren. Het kritieke gebied zijn de waarden waarin we de nulhypothese verwerpen als de steekproefgrootheid erin valt.

## Procedure van een Statistische Toets

1. Formuleer de nulhypothese (\(H_0\)) en alternatieve hypothese (\(H_1\)).
2. Bepaal het significantieniveau (\(\alpha\)).
3. Verzamel en analyseer de gegevens uit de steekproef.
4. Bereken de steekproefgrootheid en de bijbehorende p-waarde.
5. Vergelijk de p-waarde met het significantieniveau.
6. Als de p-waarde kleiner is dan \(\alpha\), verwerp dan de nulhypothese; anders, aanvaard de nulhypothese.
7. Conclusie: Besluit op basis van de testresultaten of er genoeg bewijs is om de nulhypothese te verwerpen.
