```python
import matplotlib.pyplot as plt

# Sample data for visualization
categories = ['Category A', 'Category B', 'Category C', 'Category D']
category_values = [30, 45, 20, 55]

# Bar Chart (Nominal Data)
plt.bar(categories, category_values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart for Nominal Data')
plt.show()

# Histogram (Ordinal Data)
data = [15, 25, 35, 40, 45, 50, 55, 60, 65, 70, 75]
plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram for Ordinal Data')
plt.show()

# Pie Chart (Nominal Data)
sizes = [30, 45, 20, 55]
labels = ['Category A', 'Category B', 'Category C', 'Category D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart for Nominal Data')
plt.show()

# Line Chart (Interval Data)
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
values = [100, 120, 150, 170, 200, 220, 250, 270, 300, 320, 350]
plt.plot(years, values, marker='o')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Line Chart for Interval Data')
plt.show()

# Scatter Plot (Interval Data)
x_values = [2, 3, 4, 5, 6, 7, 8]
y_values = [10, 12, 8, 15, 18, 20, 25]
plt.scatter(x_values, y_values)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot for Interval Data')
plt.show()

# Box Plot (Interval Data)
data = [12, 15, 18, 20, 22, 24, 28, 30, 35, 40, 60, 70, 80, 100]
plt.boxplot(data)
plt.ylabel('Value')
plt.title('Box Plot for Interval Data')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for visualization
x_values = [1, 2, 3, 4, 5]
y_values = [5, 10, 3, 8, 12]

# Create a scatter plot
plt.scatter(x_values, y_values, color='blue', label='Data Points')

# Mark specific points with red circles
highlighted_indices = [1, 3]
highlighted_x = [x_values[i] for i in highlighted_indices]
highlighted_y = [y_values[i] for i in highlighted_indices]
plt.scatter(highlighted_x, highlighted_y, color='red', marker='o', label='Highlighted Points')

# Add annotations with arrows
for i in highlighted_indices:
    plt.annotate(f'Point {i}', (x_values[i], y_values[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9, arrowprops=dict(facecolor='black', arrowstyle='wedge,tail_width=0.7'))

# Customize the plot
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot with Markers and Annotations')
plt.legend()

# Show the plot
plt.show()
```
